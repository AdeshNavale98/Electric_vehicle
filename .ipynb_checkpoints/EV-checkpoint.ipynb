```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Dataset
df=pd.read_csv('electric_vehicles_spec_2025.csv.csv')
df
```

```python
# Quick Preview
df.head(10)
```

```python
# information about Column or Understand Data Types & Shape
df.info()
```

```python
# Count missing value
missing=df.isna().sum()
missing
```

```python
# Generate Summary or Descriptive Statistics
df.describe()
```

```python
# Data Cleaning

#fill missing categorical data with mode
#remove duplicates
df.drop_duplicates()
```

```python
#fill missing numeric data with median

df['torque_nm']=df['torque_nm'].fillna(df['torque_nm'].median())
```

```python
df.isna().sum()
```

```python
#fill nan data with 0 in towing_capacity_kg

df['towing_capacity_kg'] = df['towing_capacity_kg'].fillna(0)
```

```python
df['number_of_cells']
```

```python
# Drop column number_of_cells because more missing values
df = df.drop(columns=['number_of_cells'])
```

```python
df['model'] = df['model'].fillna('Unknown')
```

```python
df['fast_charging_power_kw_dc']=df['fast_charging_power_kw_dc'].fillna(df['fast_charging_power_kw_dc'].median())
```

```python
df['fast_charge_port']=df['fast_charge_port'].fillna(df['fast_charge_port'].mode()[0])
```

```python
df['cargo_volume_l'] = pd.to_numeric(df['cargo_volume_l'], errors='coerce')
```

```python
df['cargo_volume_l'] = df['cargo_volume_l'].fillna(df['cargo_volume_l'].median())
```

```python
df['cargo_volume_l'].unique()
```

```python
df.isna().sum()
```

# Exploratory Data Analysis (EDA)

Correlation heatmap

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show
```

Distribution plots

```python
# Battery Capacity

sns.histplot(df['battery_capacity_kWh'],kde=True)
plt.title('Battery Capacity Distribution')
plt.show()
```

 Top Brands by Range

```python
brand_range=df.groupby('brand')['range_km'].mean().sort_values(ascending=False).head(10)
brand_range.plot(kind='barh',color='green')
plt.xlabel("Average Range(km)")
plt.title("Top Brand by Average Range")
plt.show()
```

Segment Distribution

```python
df['segment'].value_counts().plot(kind='pie',autopct='%1.0f%%',figsize=(8,8))
plt.title('Segment Share')
plt.ylabel('')
plt.show()
```

Advanced Feature Relationships

Top speed vs. range

```python
sns.scatterplot(x='top_speed_kmh',y='range_km',data=df,hue='car_body_type')
plt.title('Top Speed vs. Range by Car Body Type')
plt.show
```

Battery capacity vs. efficiency

```python
sns.scatterplot(x='battery_capacity_kWh',y='efficiency_wh_per_km',data=df,hue='segment')
plt.title('Battery Capacity vs Efficiency')
plt.show()
```

Business Insights

SUVs have the highest average towing capacity.

Models with higher battery capacity generally achieve better range, but their efficiency varies.

Fast charging power is strongly correlated with top speed.
